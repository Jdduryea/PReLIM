```python
import numpy as np
import matplotlib.pyplot as plt
import cPickle as pickle
%matplotlib inline
```

```python
hm = pickle.load(open("MacKayMouseA_atms.p"))
neurons = pickle.load(open("lm_neurons_atms.p"))
glia = pickle.load(open("lm_glia_atms.p"))
sperm = pickle.load(open("mouse_sperm_atms.p"))
```

# Model Performance

```python
# Scatter comparison
plt.scatter([x[0] for x in neurons],[x[1] for x in neurons], label="neurons")
plt.scatter([x[0] for x in glia],[x[1] for x in glia], label="glia")
plt.scatter([x[0] for x in sperm],[x[1] for x in sperm], label="sperm")

plt.xlabel("Average Accuracy")
plt.ylabel("Area Under Retention Curve")
plt.title("Model Performance")

plt.xlim([0,1])
plt.ylim([0,1])
plt.legend()
```

```python
# Scatter
plt.scatter([x[0] for x in neurons],[x[1] for x in neurons], label="neurons")
plt.scatter([x[0] for x in glia],[x[1] for x in glia], label="glia")
plt.scatter([x[0] for x in sperm],[x[1] for x in sperm], label="sperm")
plt.xlabel("Average Accuracy")
plt.ylabel("Area Under Retention Curve")
plt.title("(Zoomed in) Model Performance")

plt.xlim([0.8,1])
plt.ylim([0.5,1])
plt.legend()
```

```python
plt.boxplot([[x[0] for x in glia],[x[0] for x in neurons ], [x[0] for x in sperm ]])
plt.ylim([0,1])
plt.xticks(range(1,4),["Glia","Neuron","Sperm"])
plt.ylabel("Average Accuracy")
plt.title("Accuracy Comparisons")

plt.show()
plt.cla()

plt.boxplot([[x[0] for x in glia],[x[0] for x in neurons ], [x[0] for x in sperm ]])
plt.ylim([0.7,1])
plt.xticks(range(1,4),["Glia","Neuron","Sperm"])
plt.ylabel("Average Accuracy")
plt.title("Accuracy Comparisons (Zoomed in)")
```
